```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extracting data into Jupyter Notebook
df = pd.read_csv(r"D:\Personal\ELEVATE LABS\Task 5 29-04-2025\Titanic Dataset.csv")
```

```python
df
```

```python
# Splitting column "Name" into "FirstName" and "LastName"

df[["LastName", "FirstName"]] = df["Name"].str.split("," , expand = True)
```

```python
# Using head function

df.head()
```

```python
# Splitting column "FirstName"  into "Title" and "FirstName"

df[["Title", "FirstName"]] = df["FirstName"].str.split('.', n=1, expand=True).fillna('')
```

```python
# Using tail function

df.tail()
```

```python
# Reordering columns
# Finding the Original Position of "Name"

original_index = df.columns.get_loc("Name")
```

```python
# Dropping the "Name" Column

df.drop(columns = ["Name"], inplace = True)
```

```python
# Inserting the New Columns at the Same Position

df.insert(original_index, "Title", df.pop("Title"))
df.insert(original_index + 1, "FirstName", df.pop("FirstName"))
df.insert(original_index + 2, "LastName", df.pop("LastName"))
```

```python
# Dropping FName column

df.drop(columns = ["FName"], inplace = True)
```

```python
df.head()
```

```python
# Rounding of all fares to 2 decimals

df["Fare"] = df["Fare"].round(2)
```

```python
df.head()
```

```python
#Updating column names for consistent Format and Naming

df.columns = ['PassengerId', 'Survived', 'PClass', 'Title', 'FirstName', 'LastName',
       'Sex', 'Age', 'SibSp', 'ParCh', 'TicketNo', 'TicketFare', 'Cabin', 'Embarked']
```

```python
# Checking information of the columns

df.info()
```

```python
# Checking for any null values

df.isna().sum()
```

```python
# Filling null values in "Age" column with 0

df["Age"] = df["Age"].fillna(0)
```

```python
# Filling null values in "Cabin" column with empty string

df["Cabin"] = df["Cabin"].fillna('')
```

```python
# Filling null values in "Embarked" column with empty string

df["Embarked"] = df["Embarked"].fillna('')
```

```python
df.isna().sum()
```

```python
# Removing rows with no embarkment point

df = df[df['Embarked'] != '']
```

```python
df
```

```python
# Finding some basic statistical values from the df

df.describe().T
```

```python
# Finding survival rate by passenger class

survival_by_class = df.groupby("PClass")["Survived"].mean()

print(survival_by_class)
```

```python
# Finding survival rate by embarkation point

survival_by_embarkment = df.groupby("Embarked")["Survived"].mean()

print(survival_by_embarkment)
```

```python
# Finding survival rate by gender

survival_by_gender = df.groupby("Sex")["Survived"].mean()

print(survival_by_gender)
```

```python
# Rouding off ages to integers

df.loc[:, "Age"] = df["Age"].round().astype(int)
```

```python
df
```

```python
# Dividing ages into groups

df.loc[:, "AgeGroup"] = pd.cut(df["Age"], bins=[0, 10.5, 20, 30, 40, 50, 60, 70, 80], labels=["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71-80"])
```

```python
df.head()
```

```python
# Finding survival rate by age

survival_by_age = df.groupby("AgeGroup")["Survived"].mean()

print(survival_by_age)
```

```python
#Subplots for all above calculations

# Setting figure size for subplot
plt.figure(figsize = (15, 10))

# To display Survival Rate by Passenger Class
plt.subplot(2, 2, 1)  # Syntax: plt.subplot(no. of rows, no. of columns, position of the plot)
survival_by_class.plot(kind='bar', color=["#00008B", "#0000FF", "#4169E1"])
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survial Rate")

# To display Survival Rate by Embarkment Point
plt.subplot(2, 2, 2)
survival_by_embarkment.plot(kind='bar', color=["#006400", "#008000", "#228B22"])
plt.title("Survival Rate by Embarkment Point")
plt.ylabel("Survial Rate")

# To display Survival Rate by Gender
plt.subplot(2, 2, 3)
survival_by_gender.plot(kind='bar', color=["#8B0000", "#B22222", "#FF0000"])
plt.title("Survival Rate by Gender")
plt.ylabel("Survial Rate")

# To display HISTOGRAM
plt.subplot(2, 2, 4)
sns.histplot(x=survival_by_age.index, y=survival_by_age.values)
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Age Group")
plt.show()
```

```python
# Relation between the different variables with hues based on survival status

sns.pairplot(df, hue="Survived")  # Hue adds color based on survival status

plt.show()
```

```python
# Finding age distribution among passengers

sns.histplot(df["Age"], bins = 25, kde = True, color = "#800080")
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Finding fare distribution

plt.boxplot(df["TicketFare"], vert = False)
plt.title("Fare Distribution")
plt.show()
```

```python
passenger_count_by_embarkation = df["Embarked"].value_counts()
print(passenger_count_by_embarkation)
```

```python
# Finding passenger distribution by embarkment point

plt.pie(passenger_count_by_embarkation, labels = passenger_count_by_embarkation.index, autopct="%.2f%%", explode=[0.05, 0.05, 0.05])
plt.title("Passenger Distribution By Embarkment Point")
plt.show()
```
